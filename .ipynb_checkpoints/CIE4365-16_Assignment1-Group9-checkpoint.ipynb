```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import xlrd
%matplotlib inline
```

```python
def tic():
    #Homemade version of matlab tic and toc functions
    import time
    global startTime_for_tictoc
    startTime_for_tictoc = time.time()

def toc():
    import time
    if 'startTime_for_tictoc' in globals():
        print ("Elapsed time is " + str(time.time() - startTime_for_tictoc) + " seconds.")
    else:
        print ("Toc: start time not set")
```

```python
#**** use 'for' cyclic sentence to change the parameter value and get 'NSE' results.
#*** here change the value of S_clmax, the other nine parameters have the similar procedure.


######### The following part is to use sovle.ivp to solve the problem.
import scipy.integrate as spint


# read data from excel
meteo = pd.read_excel('meteo.xlsx')
leachate = pd.read_excel('leachate.xlsx')
Jrf = np.array(meteo['rain_station'][3451:6208:1])
pEV = np.array(meteo['pEV'][3451:6208:1])
Q_dr = np.array(leachate['leachate production']) / 28355.0

# parameters initialisation (use the mean value in Table 2)
a_cl = 0.005     # 0.005 - 0.01
a_wb = 0.00082      # 0.0005 - 0.001
b_cl = 5.0       # 0 - 80
b_wb = 30.0      # 0 - 80
S_clmax = 0.65  # 0.365-1.05
S_wbmax = 7.5     # 6 - 9.6
S_clmin = 0.0
S_wbmin = 0.0
beta_0 = 0.975    # 0 - 1
C_f = 0.94       # 0.2 - 1.2

f_red = 1.0
# initial values for S (use half of the maximum values as initial input)
S_cl0 = S_clmax / 1.547
S_wb0 = S_wbmax / 1.035


#########      Main process
# Definition of output times (0 to 2757 days)
tOut = np.linspace(0, 2757, 2757)
nOut = np.shape(tOut)[0]

# Definition of Rate Equation
def dYdt(t, S):
    return np.array([Jrf_i - a_cl * ((S[0] - S_clmin) / (S_clmax - S_clmin)) ** b_cl - pEV_i * C_f * f_red,
                     (1.0 - beta_0 * ((S[0] - S_clmin) / (S_clmax - S_clmin))) * a_cl * ((S[0] - S_clmin) / (S_clmax - S_clmin)) ** b_cl
                     - a_wb * ((S[1] - S_wbmin) / (S_wbmax - S_wbmin)) ** b_wb])

S_rslt = np.zeros([nOut, 2], dtype=float)
dtMax = 0.01
#dtMin = 1e-11
t = tOut[0]
iiOut = 0

tic()

Q_cal = np.zeros(2757)

Q_rate = np.zeros(2757)
E = np.zeros(2757)

a_plot = np.zeros([22, 1], dtype=float)
E_plot = np.zeros([22, 1], dtype=float)


for m in range (0, nOut-1):
    Q_rate[m] = Q_dr[m] - Q_dr[m-1]
Q_rate[0] =  0.
Q_ave =  np.mean(Q_rate)
print('Q_ave', Q_ave) 
print('Q_rate', Q_rate)

for k in range (0, 22):
    S_clmax = 0.4 + 0.025 * k  # change parameter values, the other nine parameters have the similar procedure
    S_cl0 = S_clmax / 1.547
    S0 = np.array([S_cl0, S_wb0]) 
    S = S0
    S_rslt[0, :] = S
    for i in range (0, nOut-1):
        t_span = [tOut[i], tOut[i+1]]
        # t_span = [0, 1]
        Jrf_i = Jrf[i]
        pEV_i = pEV[i]
        SODE = spint.solve_ivp(dYdt, t_span, S, t_eval=t_span, vectorized=True, method='RK45', rtol=1e-5)
        S = SODE.y[:, 1]
        S_rslt[i+1, :] = SODE.y[:, 1]
    SclODE = S_rslt[:, 0]
    SwbODE = S_rslt[:, 1]

    Q_cal = beta_0 * ((SclODE - S_clmin) / (S_clmax - S_clmin)) * a_cl * ((SclODE - S_clmin) / (S_clmax - S_clmin)) ** b_cl\
              + a_wb * ((SwbODE - S_wbmin) / (S_wbmax - S_wbmin)) ** b_wb
    
    #    Nash method
    E = 1 - sum((Q_rate - Q_cal) ** 2) / sum((Q_rate - Q_ave) ** 2)
    a_plot[k] = S_clmax
    E_plot[k] = E
    

print('Q_cal', Q_cal)
print('S_clmax', a_plot)
print('E', E_plot)

plt.figure()
plt.plot(a_plot, E_plot, 'b-', label='Nash' )
plt.grid()
plt.legend(loc='best')
plt.xlabel('S_clmax')
plt.ylabel('E')

plt.show()
```

```python
########      Optimization
#******** not used to determine the parameter, but can be used to examine the accuracy of parameters determined before
from scipy.optimize import minimize


Q_dr = np.array(leachate['leachate production']) / 28355.0

# Define the target funtion
def fun(args):
    Scl, Swb, Q_dr = args
    # The sequence in x is a_cl, a_wb, b_cl, b_wb, S_clmax, S_wbmax, beta_0
    v = lambda x: np.sum(np.abs(x[6] * Scl / x[4] * x[0] * (Scl / x[4]) ** x[2] + \
                         x[1] * (Swb / x[5]) ** x[3] -Q_dr))
    return v

# Define the constraint function
def con(args):
    x0min, x0max, x1min, x1max, x2min, x2max, x3min, x3max, x4min, x4max, x5min, x5max, x6min, x6max = args
    cons = ({'type': 'ineq', 'fun': lambda x: x[0] - x0min}, \
            {'type': 'ineq', 'fun': lambda x: -x[0] + x0max}, \
            {'type': 'ineq', 'fun': lambda x: x[1] - x1min}, \
            {'type': 'ineq', 'fun': lambda x: -x[1] + x1max}, \
            {'type': 'ineq', 'fun': lambda x: x[2] - x2min}, \
            {'type': 'ineq', 'fun': lambda x: -x[2] + x2max}, \
            {'type': 'ineq', 'fun': lambda x: x[3] - x3min}, \
            {'type': 'ineq', 'fun': lambda x: -x[3] + x3max}, \
            {'type': 'ineq', 'fun': lambda x: x[4] - x4min}, \
            {'type': 'ineq', 'fun': lambda x: -x[4] + x4max}, \
            {'type': 'ineq', 'fun': lambda x: x[5] - x5min}, \
            {'type': 'ineq', 'fun': lambda x: -x[5] + x5max},\
            {'type': 'ineq', 'fun': lambda x: x[6] - x6min}, \
            {'type': 'ineq', 'fun': lambda x: -x[6] + x6max})
    return cons

# Define constant parameters in the target function
args_cst = (SclODE, SwbODE, Q_rate)
# Define the minimum and maximum values of parameters to be optimzed
args_range = (5e-3, 1e-2, 5e-4, 1e-3 , 0, 80, 0, 80, 0.45, 0.975, 6, 9.6, 0, 1)
cons = con(args_range)
# Initialize the parameters
x0 = np.asarray((5e-3, 8e-4, 5, 30, 0.65, 7.5, 0.975 ))

res = minimize(fun(args_cst), x0, method='SLSQP', constraints=cons, tol=1e-6)
print(res.fun)

print(res.success)
print(res.x)

```
